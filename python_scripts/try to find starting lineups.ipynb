```python
def get_teams_trailing_10_games(inputs):
    gameID = inputs[0]
    homeTeam = inputs[1]
    awayTeam = inputs[2]
    gameDate = inputs[3]
    print (gameID)
    helper = df[df['gameDate']<gameDate]
    
    #reduce games to only the ones the team has played in the last 10 games
    away_info = helper[(helper['homeTeam']==awayTeam) | (helper['awayTeam'] == awayTeam )].tail(10)['gameID']
    home_info = helper[(helper['homeTeam']==homeTeam) | (helper['awayTeam'] == homeTeam)].tail(10)['gameID']
    away_info = huge_df[huge_df['gameID'].isin(away_info)]
    home_info = huge_df[huge_df['gameID'].isin(home_info)]

    positions = ['C','1B','2B','3B','SS','LF','CF','RF']
    global nl_teams
    nlTeams = [x.lower() for x in nl_teams]
    away_starters = []
    home_starters = []
    for pos in positions:
        try:
            temp_away = away_info[(away_info['gamePosition']==pos) &
                                     (away_info['teamID']==awayTeam)]
            temp_home = home_info[(home_info['gamePosition']==pos) & 
                                     (home_info['teamID']==homeTeam)]
            away_starters.append(temp_away['playerID'].mode()[0])
            home_starters.append(temp_home['playerID'].mode()[0])
        except IndexError as e:
            print (f'error in {gameID}. the {homeTeam} vs. {awayTeam} at {pos}')
            return []
    
    extra_position = None
    if homeTeam in nlTeams:
        if homeTeam == 'hou' and gameDate.year>=2013:
            extra_position = 'DH'
        else:
            # If it's an NL game, the "expected starter" is not found from previous 10 games
            # Therefore, just return it as is 
            final_output = []
            for i in range(8):
                final_output.append([gameID,away_starters[i],awayTeam])
                final_output.append([gameID,home_starters[i],homeTeam])
            return final_output
    else:
        extra_position = 'DH'
    
    try:
        temp_away = away_info[(away_info['gamePosition']==extra_position) &
                             (away_info['teamID']==awayTeam)]
        temp_home = home_info[(home_info['gamePosition']==extra_position) &
                             (away_info['teamID']==homeTeam)]
        away_starters.append(temp_away['playerID'].mode()[0])
        home_starters.append(temp_home['playerID'].mode()[0])
    except IndexError as e:
        print (f'error in {gameID}. the {homeTeam} vs. {awayTeam}. Unable to get {extra_position}')
        final_output = []
        for i in range(8):
            final_output.append([gameID,away_starters[i],awayTeam])
            final_output.append([gameID,home_starters[i],homeTeam])
        return final_output
    
    final_output = []
    for i in range(9):
        final_output.append([gameID,away_starters[i],awayTeam])
        final_output.append([gameID,home_starters[i],homeTeam])
    
    
    return final_output
```
